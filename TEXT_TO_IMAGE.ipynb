```python
# Install the necessary libraries
!pip install diffusers transformers torch

# Import the required libraries
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image

# Load the Stable Diffusion model
model_id = "runwayml/stable-diffusion-v1-5"  # Updated model ID
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to("cuda")  # Move the model to GPU for faster inference (if available)

# Function to generate image from text
def generate_image(prompt):
    # Generate the image
    image = pipe(prompt).images[0]
    return image

# Example: Generate an image from a textual description
prompt = "a futuristic city skyline during sunset with flying cars"
image = generate_image(prompt)

# Display the generated image
image.show()

# Save the generated image
image.save("generated_image.png")
```
